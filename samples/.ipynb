```python
import quandl
```

```python
df = quandl.get('WIKI/GOOGL')
```

```python
df
```

```python
df_used = df[['Adj. Open','Adj. High','Adj. Low','Adj. Close','Adj. Volume']]
```

```python
df_used.head()
```

```python
df_used['HL_PCT'] = (df_used['Adj. High'] - df_used['Adj. Close']) / df_used['Adj. Close']*100.0
```

```python
df_used['PCT_change'] = (df_used['Adj. Close'] - df_used['Adj. Open']) / df_used['Adj. Open']*100.0
```
